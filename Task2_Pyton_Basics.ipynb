```python
# 1. String - Text data
name = "Hashir"
print("String Example")
print("Value:", name)
print("Type:", type(name))
print("Lenght:", len(name))
```

```python
# 2. Integer - Whole number
age = 23
print("Integer Example")
print("Value:", age)
print("Type:", type(age))
print("Age + 5:", age + 5)
```

```python
# 3. Float - Decimal number
pi = 3.14159
price = 99.99
print("Float Example")
print("pi:", pi, "Type:", type(pi))
print("price:", price, "Type:", type(price))
```

```python
# 4. Boolean - True/False
is_student = True
has_completed = False
print("Boolean Example")
print("Is student:", is_student, "Type:", type(is_student))
print("Completed:", has_completed)
```

```python
# 5. List - Ordered collection
marks = [85, 95, 74, 98, 55]
print("List Example")
print("Marks:", marks)
print("type:", type(marks))
print("First Marks:", marks[0])
print("Total Marks:", sum(marks))
```

```python
# 6. Tuple - Immutable coleection
coordinates = (31.5204, 74.3583) # Karachi lat, long
print("Tuple Example")
print("Coordinates:", coordinates)
print("Type:", type(coordinates))
print("Latitude:", coordinates[0])
```

```python
# 7. Dictionary - key-Value pairs
student = {"name": "Hashir", "age": 21, "city": "Daharki", "marks": 85}
print("Dictionary Example")
print("Student:", student)
print("Type:", type(student))
print("Name:", student["name"])
print("Keys:", student.keys())
```

```python
# 8. For Loop
print("For Loop Example - 1 to 5:")
for i in range(1, 6):
  print("Number:", i)
```

```python
# 9. While Loop
print("While Loop Example - Countdown")
count = 5
while count > 0:
  print(count)
  count = count - 1
  print("Done!")
```

```python
# 10. If-Else Conditional
marks = 75
print("If-Else Example")
if marks >= 80:
  print("Grade: A")
elif marks >= 60:
  print("Grade: B")
else:
  print("Grade: C")
print(":Your marks:", marks)
```

```python
# 11. Function with return
def add(a, b):
  result = a + b
  return result
print("function Example")
print("5 + 3 =", add(5, 3))
print("10 + 20 =", add(10, 20))
```

```python
# 12. Function with default parameter
def greet(name="Student"):
  return "Hello " + name
print("Function with Default")
print(greet())
print(greet("Hashir"))
```

```python
# 13. List Comprehension
squares = [x*x for x in range(1, 6)]
evens = [x for x in range(10) if x % 2 == 0]
print("List Comprehension Example")
print("Squares 1-5:", squares)
print("Even number 0-9:", evens)
```

```python
# 14. NumPy Array
import numpy as np
arr = np.array([10, 20, 30, 40, 50])
print("NumPy Array Example")
print("Array:", arr)
print("Mean:", arr.mean())
print("Sum:", arr.sum())
print("Max:", arr.max())
```

```python
# 15. NumPy Math Operations
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("NumPy Operations")
print("a:", a)
print("b:", b)
print("a + b:", a + b)
print("a * b:", a * b)
```

```python
# 16. Pandas + Sample Dataset from Dictionary
import pandas as pd
data = {
    "Name": ["Hashir", "Alishba", "Ahmed", "Fatima", "Ali"],
    "Age": [23, 21, 27, 19, 22],
    "City": ["Daharki", "Karachi", "Lahore", "Sukkur", "Nawanshah"],
    "Marks": [92, 85, 71, 81, 77]
}
df = pd.DataFrame(data)
print("Pandas DataFrame Example")
print("Dataset:")
print(df)
print("\nBasic Info:")
print(df.info())
print("\nStatistics:")
print(df.describe())
print("\nFirst 3 rows:")
print(df.head(3))
```
